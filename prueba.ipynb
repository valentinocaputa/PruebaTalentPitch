```python
#Convierto la columna updated_at al formato de fecha y extraigo 
df_users['updated_at'] = pd.to_datetime(df_users['updated_at'])
df_users['updated_at'] = df_users['updated_at'].dt.date
```

```python
df_users.head()
```

```python
#Separo las columnas con listas en nuevas columnas

df_users['want_to_work'] = df_users['want_to_work'].str.replace('[', '').str.replace(']', '').str.replace("'", "")
df_users['benefits_in_companies'] = df_users['benefits_in_companies'].str.replace('[', '').str.replace(']', '').str.replace("'", "")
```

```python
#Reemplazo guiones bajos con espacios en blanco 
df_users['want_to_be'] = df_users['want_to_be'].str.replace('_', ' ')
df_users['type'] = df_users['type'].str.replace('_', ' ')
```

```python
#Extraigo el nombre en español de un objeto tipop json
def extract_name(json_str):
    try:
        name_dict = json.loads(json_str)
        return name_dict.get('es')
    except (json.JSONDecodeError, TypeError):
        return None

#Aplicar la funcion para extraer el nombre en español
df_users['name'] = df_users['name'].apply(extract_name)
```

#### Destacar beneficios laborales atractivos

```python
#Frecuencia de los beneficios laborales atractivos
benefits_freq = df_users['benefits_in_companies'].str.split(', ').explode().value_counts()

print("\nFrecuencia de beneficios laborales:")
print(benefits_freq)
```
